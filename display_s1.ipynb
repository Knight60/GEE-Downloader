# แสดงผลภาพ Sentinel-1 ที่ดาวน์โหลดมา
โค้ดส่วนนี้ใช้ `rasterio` และ `matplotlib` ในการอ่านไฟล์ `.tif` ที่มี 3 แบนด์ (Multi-Temporal) มาแสดงผลเป็นภาพสี RGB เพื่อดูการเปลี่ยนแปลงของพื้นที่ในช่วงเวลาที่กำหนด

```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np
import os

desc_path = 'download/desc_change.tif'
asc_path = 'download/asc_change.tif'
```

```python
def plot_s1(image_path, title):
    if not os.path.exists(image_path):
        print(f"ไม่พบไฟล์: {image_path} กรุณาตรวจสอบว่าดาวน์โหลดเสร็จสิ้นแล้ว")
        return
    
    with rasterio.open(image_path) as src:
        # อ่านข้อมูลภาพทั้ง 3 แบนด์ (spring, late_spring, summer)
        img = src.read()
        
        # ค่าของ Sentinel-1 (VV) มักจะอยู่ในช่วง -25 ถึง 5 dB
        # ทำการปรับค่า (Normalize) ให้อยู่ในช่วง 0-1 สำหรับการแสดงผล RGB
        vmin, vmax = -25, 5
        img_norm = (img - vmin) / (vmax - vmin)
        img_norm = np.clip(img_norm, 0, 1)
        
        # matplotlib ต้องการรูปแบบข้อมูลแบบ (height, width, channels) 
        # แต่ rasterio คืนค่ามาเป็น (channels, height, width)
        img_rgb = np.transpose(img_norm, (1, 2, 0))
        
        # แสดงผลภาพ
        plt.figure(figsize=(12, 10))
        plt.imshow(img_rgb)
        plt.title(title, fontsize=16)
        plt.axis('off')
        plt.show()
```

```python
# แสดงภาพ DESCENDING
plot_s1(desc_path, 'Sentinel-1 DESCENDING (Multi-Temporal RGB)')
```

```python
# แสดงภาพ ASCENDING
plot_s1(asc_path, 'Sentinel-1 ASCENDING (Multi-Temporal RGB)')
```
